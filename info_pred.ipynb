```python
from sklearn.preprocessing  import MinMaxScaler
```

```python
normalizer = MinMaxScaler(feature_range=(0,1))
```

```python
profit_2 =  normalizer.fit_transform(price_1)
```

```python
def datagen(data , steps):
    X_t=[]
    y_t=[]
    for i in range(len(data)-steps) :
        X_t.append(data[i:(i+steps)])
        y_t.append(data[i+steps])

    return X_t , y_t    
        
```

```python
price_train= profit_2[0:3970]
```

```python
price_test = profit_2[3970:]
```

```python
price_train[10]
```

```python
X_train , y_train = datagen(price_train , 10)
```

```python
X_train[10]
```

```python
y_train = np.array(y_train)
```

```python
X_train = np.array(X_train)
```

```python
X_train.shape
```

```python
from keras.models import Sequential
```

```python
model = Sequential()
```

```python
from keras.layers import LSTM
```

```python
from keras.layers import Dense
```

```python
model.add(LSTM(units= 512 , return_sequences = True ))
```

```python
model.add(LSTM(units= 256))
```

```python
model.add(Dense(units=128 , activation = 'relu'))
```

```python
model.add(Dense(units=1 , activation = 'linear'))
```

```python
model.compile(optimizer='rmsprop' , loss='mean_squared_error')
```

```python
model.get_config()
```

```python
model.fit(X_train,y_train ,epochs=50)
```

```python
model.history.history
```
