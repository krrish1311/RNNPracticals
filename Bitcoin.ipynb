```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("bitcoin_csv.csv")
```

```python
dataset
```

```python
price = dataset["price(USD)"]
```

```python
price = price.values
```

```python
null= price.isnull()
```

```python
def datagen(data ,null, steps) :
    X=[]
    y=[]
    for i in range(len(data) - steps):
        if null[i]:
            print('error')
            
        else :
            X.append(data[i:i+steps])
            y.append(data[i+steps])

       
    return X , y                      
```

```python
X , y = datagen(price ,null , 16)
```

```python
y_arr = np.array(y)
```

```python
X_arr = np.array(X)
```

```python
price
```

```python
X_train = X_arr[0:1500]
y_train = y_arr[0:1500]
```

```python
X_test = X_arr[1500:]
y_test = y_arr[1500:]
```

```python
X_arr.shape
```

```python
X_train_final = np.expand_dims(X_train , axis = 0)
```

```python
X_train_final = X_train_final.reshape(1500,16,1)
```

```python
X_train_final.shape
```

```python
y_train_final = y_train.reshape(-1,1)
```

```python
from keras.models import Sequential
model = Sequential()
```

```python
from keras.layers import LSTM 
from keras.layers import Dense
```

```python
model.add(LSTM(units = 64 , return_sequences=True))
```

```python
model.add(LSTM(units = 32))
```

```python
model.add(Dense(units = 124 , activation = 'relu' ))
```

```python
model.add(Dense(units = 68 , activation = 'relu' ))
```

```python
model.add(Dense(units = 1 , activation = 'linear' ))
```

```python
model.compile(optimizer='rmsprop' , loss='mean_squared_error')
```

```python
model.fit(X_train_final,y_train_final , epochs = 100)
```

```python
X_test_final = np.expand_dims(X_test , axis = 0)
```
